Objetivo: Administrar un inventario de una tienda

Inventario inicial:
```
inventario = [
    {"producto": "camisa", "precio": 25900, "stock": "11"},
    {"producto": "pantalón", "precio": 39900, "stock": "23"},
]
```

1. Añade un nuevo producto al inventario con las siguientes caracteristicas:
 - {"producto": "abrigo", "precio": 50000, "stock": "2"}
2. Convierte los valores del stock en numeros enteros
3. Actualiza el precio de cada producto sumando 10000 al precio original
4. Imprime cada producto del inventario y sus caracteristicas con el siguiente formato:
```
print(f"Hay {stock} unidades del producto {producto}. Su precio por unidad es de {precio}")
```


````python
# Inventario inicial:
inventario = [
    {"producto": "camisa", "precio": 25900, "stock": "11"},
    {"producto": "pantalon", "precio": 39900, "stock": "23"},
]
````

````python
nuevo_producto = {
    "producto": "abrigo",
    "precio": 50000,
    "stock": "2"}

inventario.append(nuevo_producto)
````

````python
inventario
````

````python
def cambiar_stock(diccionario):
  diccionario["stock"] = int(diccionario.get("stock"))
  return diccionario

nuevo_inventario = list(map(cambiar_stock,inventario))
nuevo_inventario
````

````python
# 1. acceda al inventario (recorrer la lista)
# 2. acceda al dicionario que hay en cada pocision
# 3. tome la llave stock y conviertala a entero
# 4. reasigne el nuevo valor (entero) a la llave de stock para que se actualice

for producto in inventario:
  producto["stock"] = int(producto.get("stock"))
  producto["precio"] = producto["precio"] + 10000
  stock = producto['stock']
  prod = producto['producto']
  precio = producto['precio']
  print(f"Hay {stock} unidades del producto {prod}. Su precio por unidad es de {precio}")
````
