2. ¿Cuál es la suma total de los valores del arreglo?

4. Crea una matriz 10x10 de enteros aleatorios (0-9).

```python
# Arreglo 5x3 de números aleatorios entre 0 y 100
import numpy as np

arreglo = np.random.randint(0, 101, size=(5, 3))
print(arreglo)
```

```python
# La suma total de los valores del arreglo
suma_total = np.sum(arreglo)
print(suma_total)
```

```python
# promedio de cada columna del arreglo
promedio_columnas = np.mean(arreglo, axis=0)
print(promedio_columnas)
```

```python
# Matriz 10x10 de enteros aleatorios (0-9)
matriz = np.random.randint(0, 10, size=(10, 10))
print(matriz)
```

```python
# -1 todos los valores > 5
matriz[matriz > 5] = -1
print(matriz)
```
